```python
zoom_ascent = np.copy(ascent[200:400,350:])
print(zoom_ascent)
```

```python
plt.imshow(zoom_ascent)
```

## Augmentons le contrasque

```python
zoom_ascent[zoom_ascent > 150] = 255
plt.imshow(zoom_ascent)
```

```python
zoom_ascent[zoom_ascent < 100] = 0
plt.imshow(zoom_ascent)
```
